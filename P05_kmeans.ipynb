We can plot the heterogeneity for k's.

Since heterogeneity is the sum of variances of clusters of each labels.

As the number of labels increase, the variance of each cluster will drop, becoming 0 when the numbers of labels equal the numbers of total data points.

However, there will be an elbow in H/k graph, at which point k is optimal.

```python
from P05 import plot_heterogeneity
hetero_list = plot_heterogeneity(x1, x2)
```
